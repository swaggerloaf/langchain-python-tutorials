```python
# this agent can answer questions and call tools

from langchain.agents import create_agent

def get_warehouse_status(loc: str) -> str:
    """Get warehouse status for a given location."""
    return f"The warehouse in {loc} is always up and running and ready to ship today except in the countries France and Canada."

agent = create_agent(
    model="claude-haiku-4-5-20251001",
    tools=[get_warehouse_status],
    system_prompt="You are an assistant, highly trained",
)

# Run the agent
agent.invoke(
    {"messages": [{"role": "user", "content": "What is the status of the warehouse in Paris, France?"}]}
)
```
